```python
import pandas as pd
```

```python
# 파일들로부터 데이터 읽어오기
file_path = 'https://raw.githubusercontent.com/parkjy0629/tensorflow_study/main/study_csv/lemonade.csv'
lemonade = pd.read_csv(file_path)

file_path = 'https://raw.githubusercontent.com/parkjy0629/tensorflow_study/main/study_csv/boston.csv'
boston = pd.read_csv(file_path)

file_path = 'https://raw.githubusercontent.com/parkjy0629/tensorflow_study/main/study_csv/iris.csv'
iris = pd.read_csv(file_path)
```

```python
# 데이터의 모양확인
print(lemonade.shape)
print(boston.shape)
print(iris.shape)
```

```python
# 데이터 컬럼확인
print(lemonade.columns)
print(boston.columns)
print(iris.columns)
```

```python
# 독립변수와 종속변수 분리
i_var = lemonade[['온도']]
d_var = lemonade[['판매량']]
print(i_var.shape, d_var.shape)

i_var = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]
d_var = boston[['medv']]
print(i_var.shape, d_var.shape)

i_var = iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
d_var = iris[['품종']]
print(i_var.shape, d_var.shape)
```
