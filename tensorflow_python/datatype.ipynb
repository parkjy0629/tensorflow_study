```python
# 라이브러리 사용
import pandas as pd
```

```python
# 파일 읽어오기
file_path = 'https://raw.githubusercontent.com/parkjy0629/tensorflow_study/main/study_csv/iris2.csv'
iris = pd.read_csv(file_path)
iris.head()
```

```python
# 원핫인코딩
icd = pd.get_dummies(iris)
icd.head()
```

```python
# 데이터 타입 확인
print(iris.dtypes)
```

```python
# 품종 타입을 범주형으로 변경
iris['품종'] = iris['품종'].astype('category')
print(iris.dtypes)
```
