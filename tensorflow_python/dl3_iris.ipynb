```python
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```python
# 1. 과거 데이터 준비
file_path = 'https://raw.githubusercontent.com/parkjy0629/tensorflow_study/main/study_csv/iris.csv'
iris = pd.read_csv(file_path)
iris.head()
```

```python
# 2. 원핫인코딩
inc = pd.get_dummies(iris)
inc.head()
```

```python
print(inc.columns)
```

```python
# 3. 독립변수, 종속변수 분리
i_var = inc[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
d_var = inc[['품종_setosa', '품종_versicolor', '품종_virginica']]
print(i_var.shape, d_var.shape)
```

```python
# 4. 모델 구조 만들기
X = tf.keras.layers.Input(shape=[4])
Y = tf.keras.layers.Dense(3, activation='softmax')(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# 5. 모델 학습
model.fit(i_var, d_var, epochs=100)
```

```python
# 모델 이용
model.predict(i_var[0:5])
```
