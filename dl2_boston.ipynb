```python
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```python
# 1. 과거 데이터 준비
file_path = 'https://raw.githubusercontent.com/parkjy0629/tensorflow_study/main/study_csv/boston.csv'
file_path2 = '/content/boston.csv'
boston = pd.read_csv(file_path)
boston2 = pd.read_csv(file_path2)
print(boston.columns)
boston.head()
print(boston2.columns)
boston2.head()
```

```python
# 2. 독립변수, 종속변수 분리
i_var = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]
d_var = boston[['medv']]
print(i_var.shape, d_var.shape)
```

```python
# 3. 모델 구조 만들기
X = tf.keras.layers.Input(shape=[13])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
# 4. 데이터 모델 학습
model.fit(i_var, d_var, epochs=10)
```

```python
model.fit(i_var, d_var, epochs=100)
```

```python
model.fit(i_var, d_var, epochs=100)
```

```python
model.fit(i_var, d_var, epochs=1000, verbose=0)
```

```python
# 5. 모델 이용
model.predict(i_var[0:5])
```
